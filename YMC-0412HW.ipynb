```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def lossfunction(x):
    return  x**2 + 2*x + 1
def dlossfunction(x):
    return 2*x + 2 
x=np.linspace(-5,5,1000)
plt.plot(x,lossfunction(x))
```

```python
n=random.uniform(-5, 5) #隨機權重
print(n)
```

```python
def gradient(x):
    x = x - lr * dlossfunction(x)
    return x

lr = 0.1
x=np.linspace(-5,5,1000)
plt.plot(x,lossfunction(x))
random.uniform(-30, 30) #隨機權重
for i in range(3000): #迭代次數
    x=gradient(x)
    plt.scatter(x,lossfunction(x))
```
